# benchcheck on a GPU (Colab)

Runs the LLM benchmark-contamination detector against a **modern open-weight model** (Qwen2.5-3B by default) across four public benchmarks, using a free T4 GPU.

**Before running:** Runtime → Change runtime type → **T4 GPU**.

Takes a few minutes. Results download as JSON at the end.

## 1. Confirm GPU

```python
!nvidia-smi -L || echo 'No GPU -- set Runtime > Change runtime type > T4 GPU'
```

## 2. Clone the repo and install dependencies

```python
!git clone https://github.com/selinaaavt/benchcheck.git
%cd benchcheck
!pip install -q numpy scipy transformers datasets
# torch is preinstalled on Colab with CUDA.
```

## 3. Fetch the benchmarks (converted to benchcheck JSONL)

```python
for b in ['arc_easy','arc_challenge','openbookqa','sciq']:
    !python -m examples.fetch_benchmark --benchmark {b} --limit 300 --out examples/{b}.jsonl
```

## 4. Run the detector on a modern model

Qwen2.5-3B (Apache-2.0, late 2024). On a T4 this runs all four checks across all four benchmarks in a few minutes. Swap `--model` for any causal LM you can fit (e.g. `meta-llama/Llama-3.2-3B` if you've accepted its license + set an HF token).

```python
!python -m examples.colab_run --model Qwen/Qwen2.5-3B --limit 300
```

## 5. Download the result JSONs

Drop these into the repo's `examples/results/` to update the writeup.

```python
from google.colab import files
import glob
for f in glob.glob('results/*.json'):
    print('downloading', f)
    files.download(f)
```
